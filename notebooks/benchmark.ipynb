```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from spinup import ppo,vpg
import tensorflow as tf
import gym
from time import time,sleep
sns.set_style("darkgrid")
```

# BENCHMARK ALGO

```python
envs = ['MountainCarContinuous-v0','CartPole-v0','LunarLanderContinuous-v2']
algs = [ppo,vpg]
alg_names = ['ppo','vpg']
layers = [[32],[64,32],[64,32,16]]
reads = {}
times = {}
max_means = {}
```

```python
%%capture
for env in envs:
    reads[env] = {}
    times[env] = {}
    max_means[env] = []
    for i,alg in enumerate(algs):
        print(i)
        for l,layer in enumerate(layers):
            tf.reset_default_graph()
            
            env_fn = lambda : gym.make(env)
            
            ac_kwargs = dict(hidden_sizes=layer, activation=tf.nn.relu)
            
            logger_kwargs = dict(output_dir='data/'+alg_names[i]+'_'+env, exp_name='experiment_name')
            
            start = time() 
            alg(env_fn=env_fn, ac_kwargs=ac_kwargs, steps_per_epoch=3000, epochs=500, logger_kwargs=logger_kwargs, save_freq=50)
            reads[env][alg_names[i]+str(l)] = pd.read_csv('data/'+alg_names[i]+'_'+env+'/progress.txt',sep='\t')
            times[env][alg_names[i]+str(l)] = time() - start
            max_means[env].append({'alg_name':alg_names[i]+str(l),
                                               'time': times[env][alg_names[i]+str(l)],
                                               'max':reads[env][alg_names[i]+str(l)]['AverageEpRet'].max()})
```

```python
cols = {'ppo0':'aqua',
    'ppo1':'blue',
    'ppo2':'navy',
    'vpg0':'lightcoral',
    'vpg1':'orange',
    'vpg2':'tomato',
}
```

```python
cols
```

```python
for env in envs:
    fig,ax = plt.subplots(figsize=(10,10))
    for i,alg_name in enumerate(alg_names):
        for l,layer in enumerate(layers):
            ax.fill_between(reads[env][alg_name+str(l)]['Epoch'],reads[env][alg_name+str(l)].MinEpRet, reads[env][alg_name+str(l)].AverageEpRet, alpha=0.2,color =cols[alg_name+str(l)] )
            ax.fill_between(reads[env][alg_name+str(l)]['Epoch'],reads[env][alg_name+str(l)].MaxEpRet, reads[env][alg_name+str(l)].AverageEpRet, alpha=0.2,color = cols[alg_name+str(l)])
            ax.plot(reads[env][alg_name+str(l)]['Epoch'],reads[env][alg_name+str(l)]['AverageEpRet'],cols[alg_name+str(l)],label=alg_name+str(l))
    ax.set_title("cumulative reward (mean,min,max) per epoch in: "+env)
    ax.legend()
```

```python
for key in envs:
    obj = max_means[key]
    tmp = pd.DataFrame(obj)
    tmp = tmp.sort_values(by=['max'])
    fig,ax = plt.subplots(figsize=(10,10))
    #sns.barplot(x="max", y="time", palette = 'alg_name', data=tips) 
    sns.barplot(x="max", y="time", data=tmp) 
    ax.set_title("max mean cumulative reward by time in: "+env)
```
