```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import gym
from time import time,sleep
sns.set_style("darkgrid")
```

```python
from spinup import vpg #Vanilla Policy Gradient
from spinup import ppo #Proximal Policy Optimization

from spinup import trpo #Trust Region Policy Optimization
from spinup import sac #Soft Actor-Critic
```

# Train agent for an environment

```python
env = 'LunarLanderContinuous-v2'
alg = ppo


layers = [63,32,16]

file_name = 'lander'
```

```python
tf.reset_default_graph()
            
env_fn = lambda : gym.make(env)
            
ac_kwargs = dict(hidden_sizes=layers, activation=tf.nn.relu)
            
logger_kwargs = dict(output_dir='data/' + file_name, exp_name='experiment_name')
            
alg(env_fn=env_fn, ac_kwargs=ac_kwargs, steps_per_epoch=10000, epochs=250, logger_kwargs=logger_kwargs)
reads = pd.read_csv('data/'+file_name+'/progress.txt',sep='\t')
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.fill_between(reads['Epoch'],reads.MinEpRet, reads.AverageEpRet, alpha=0.2,color ='blue' )
ax.fill_between(reads['Epoch'],reads.MaxEpRet, reads.AverageEpRet, alpha=0.2,color = 'blue')
ax.plot(reads['Epoch'],reads['AverageEpRet'])
ax.set_title("cumulative reward (mean,min,max) per epoch")
```

# Test agent for this environment

```python
env = env_fn()
file_name = file_name
n_test = 100
print_ = 0
```

```python
#load model
from spinup.utils.test_policy import load_policy, run_policy
_, get_action = load_policy('data/' + file_name)
```

```python
state = env.reset()
rews = np.zeros(n_test)
for _ in range(n_test):
    action = get_action(state)
    state, reward, done, info = env.step(action)
    if print_:
        print('action: ')
        print(action)
        print('state: ')
        print(state)
        print('reward: ')
        print(reward)
        print()
    rews[_] = reward
    if done:
        print('done')
        state = env.reset()
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.plot(rews)
ax.set_title("reward test")
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.plot(rews.cumsum())
ax.set_title("cumulative reward test")
```
